```python
#Question 3
A = [77, 150, 210, 125, 38]
Total = np.sum(A)
print(Total,'is the total no. of students')
s_0 = 0.15*Total
s_1 = 0.24*Total
s_2 = 0.38*Total
s_3 = 0.18*Total
s_4 = 0.05*Total

#degree of freedom = n-1 = 4
#critical vaalue at 5% significance level = 9.488
#critical vaalue at 10% significance level= 7.779

chi_2 = 0


chi_2 = ((A[0]-s_0)**2)/s_0 + ((A[1]-s_1)**2)/s_1 + ((A[2]-s_2)**2)/s_2 + ((A[3]-s_3)**2)/s_3 + ((A[4]-s_4)**2)/s_4
print(chi_2,'is normal at 5% significance level but not at 10% significance level')
```

```python
#Question 4


shipment_A = [4.65, 4.84, 4.59, 4.75, 4.63, 4.75, 4.58, 4.82, 4.86, 4.60, 4.77, 4.65,4.80]
shipment_B = [4.75, 4.79, 4.74, 4.74, 4.77, 4.58, 4.81]
mean_shipment_A = np.mean(shipment_A)
print(mean_shipment_A,'is the mean for shipment A')
mean_shipment_B = np.mean(shipment_B)
print(mean_shipment_B,'is the mean for shipment B')

variance_A = np.var(shipment_A)
print(variance_A,'is the variance for shipment A')
variance_B = np.var(shipment_B)
print(variance_B,'is the variance for shipment B')

f_test = variance_A/variance_B
print(f_test,'< 2.287 therefore they belong to same group')

t_test = (mean_shipment_A-mean_shipment_B)/np.sqrt(((variance_A/13)+(variance_B/7)))

print(abs(t_test),'< 2.201 therefore they belong to same group')
```
